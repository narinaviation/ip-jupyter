```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from math import ceil, floor, pow, log

%matplotlib notebook
```

```python
img_path = 'image.jpg'
```

```python
img_raw = cv2.imread(img_path)
img = cv2.resize(src=img_raw,dsize=(400,240))
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

print('shape', img_gray.shape)
print('dtype', img_gray.dtype)
```

```python
plt.figure(1)
plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.subplot(1,2,2)
plt.imshow(img_hsv)
print("White Area : ",img_hsv[221,373])
print("Black Area : ",img_hsv[203,363])
```

```python
intensity1 = 150
intensity2 = 100

img_output = np.zeros((240,400))
for x in range(0,img_output.shape[0]):
    for y in range(0,img_output.shape[1]//2):
        img_output[x][y] = img_hsv[x][y][2] > intensity1
        
for x in range(0,img_output.shape[0]):
    for y in range(img_output.shape[1]//2,img_output.shape[1]):
        img_output[x][y] = img_hsv[x][y][2] > intensity2
        
plt.figure(2)
plt.imshow(img_output,'gray')
```

```python
pts1 = np.float32([[75, 150], [75, 255], [240, 55], [240, 380]])
pts2 = np.float32([[0, 0], [img_output.shape[0] , 0], [0, img_output.shape[1]], [img_output.shape[0], img_output.shape[1]]])

matrix = cv2.getPerspectiveTransform(pts1, pts2)
result = cv2.warpPerspective(img_output, matrix, (img_output.shape[0], img_output.shape[1]))
```
