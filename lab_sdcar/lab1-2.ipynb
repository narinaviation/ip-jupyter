```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
file_path = 'thai.jpg'
```

```python
img = cv2.imread(file_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_ycbcr = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
lower_blue = np.array([10,100,180])
upper_blue = np.array([20,255,255])

mask = cv2.inRange(img_hsv, lower_blue, upper_blue)
result_mask = cv2.bitwise_and(img_hsv, img_hsv, mask=mask)
```

```python
result = cv2.cvtColor(result_mask, cv2.COLOR_HSV2RGB)
img_roi_rgb = cv2.addWeighted(result, 1, img_rgb, 0.2, 0)
```

```python
plt.subplot(2,2,1), plt.imshow(mask)
plt.subplot(2,2,2), plt.imshow(result_mask)
plt.subplot(2,2,3), plt.imshow(result)
plt.subplot(2,2,4), plt.imshow(img_roi_rgb)
```

```python
plt.figure('Figure 4')
plt.subplot(2, 2, 1), plt.imshow(img_rgb)
plt.subplot(2, 2, 2), plt.imshow(img_roi_rgb)
plt.subplot(2, 2, 3), plt.imshow(img_ycbcr)
plt.subplot(2, 2, 4), plt.imshow(img_hsv)
```
